```python
import numpy as np
from scipy.special import factorial
import pandas as pd
```

```python
def combinatorio(N, R):
    n = factorial(N)
    r = factorial(R)
    t = factorial((N - R))
    
    return int(n / (r*t))
```

```python
combinatorio(20+1+(2*1), 20)
```

```python
def somatorio(k, valor_um, valor_dois):
    soma = 0
    for i in range(k):
        soma += combinatorio(valor_um, valor_dois)
    return soma
```

```python
def calculo_combinatorio_simplex(r, variacao, iteracao):
    
    s = 1
    k = 0
    j = 1
    funcao_um = 0
    funcao_dois = 0
    funcao_tres = 0
    funcao_quatro = 0
    maximo = 0
    
    erres = []
    esses = []
    kas = []
    maximos = []
    valor_um = []
    valor_dois = []
    valor_tres = []
    valor_quatro = []
    
    while(j <= variacao):
        # Função que é constante, dependendo somente de r
        funcao_um = combinatorio(3*r -2, r)
        funcao_quatro = k*(2*r-1)
        
        # Função que varia com base em k
        funcao_tres = k * combinatorio(2*r -1, r)
        
        # Função que é somatório com base em k
        valor_da_funcao = 0
        i = 0
        funcao_dois = 0
        while(i <= k):
            funcao_dois += combinatorio((r+s+(2*i)), r)
            i+=1
        valor_da_funcao += funcao_dois
        
        # Valor Maximo
        maximo = np.max([funcao_dois, funcao_tres, funcao_quatro])
        
        # Salva os dados
        if(funcao_um > maximo):
            maximos.append(1)
        else:
            maximos.append(0)
        erres.append(r)
        esses.append(s)
        kas.append(k)
        valor_um.append(funcao_um)
        valor_dois.append(valor_da_funcao)
        valor_tres.append(funcao_tres)
        valor_quatro.append(funcao_quatro)
        
        # Iteração
        j += 1

        # Auxilio da iteração
        if(k == iteracao):
            k = 0
            if(s == 2):
                s = 1
                if(r == 1):
                    r += 4
                else:
                    r += 5
            else:
                s += 1
        else:
            k += 1
        
    
    # Dados para DataFrame e Excel
    frame = pd.DataFrame()
    frame["r"] = erres
    frame["s"] = esses
    frame["k"] = kas
    frame["1: (3r-2, r)"] = valor_um
    frame["2: SUM((r+s+2i, r))"] = valor_dois
    frame["3: k(2r-1, r)"] = valor_tres
    frame["4: k(2r-1)"] = valor_quatro
    frame["Maior: 1 > max(2, 3, 4)"] = maximos
     
    
    return frame
```

```python
df = calculo_combinatorio_simplex(10, 252, 20)
```
